```python
from pymongo import MongoClient
import pandas as pd
# Connect to MongoDB using PySpark
from pyspark.sql import SparkSession
```

### Create a SpartContent within a SparkSession to connect to MongoDB

```python

# Asegúrate de reemplazar "3.0.1" con la versión correcta del conector de MongoDB para tu versión de Spark/Scala
spark_mongo = SparkSession.builder \
    .appName("PySpark MongoDB App") \
    .config("spark.executor.memory", "2g") \
    .config("spark.sql.shuffle.partitions", "2") \
    .config("spark.mongodb.input.uri", "mongodb://localhost:27017/Contrataciones.Sample_Contratos_EDCA") \
    .config("spark.mongodb.output.uri", "mongodb://localhost:27017/Contrataciones.Sample_Contratos_EDCA") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.2.4") \
    .getOrCreate()

sc = spark_mongo.sparkContext
```
